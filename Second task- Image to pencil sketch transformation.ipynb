# Image to grayscale image

Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

Loading image from desktop

```python
img = mpimg.imread(r'C:\Users\WIN7\Pictures\rose.jpg')
```

Converting it into gray scale picture

```python
gray = rgb2gray(img)
```

```python
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

# Normal image to blurred image

importing library

```python
from PIL import Image, ImageFilter
```

Original image

```python
OriImage = Image.open(r'C:\Users\WIN7\Pictures\rose.jpg')
OriImage.show()
```

Blurring the image

```python
blurImage = OriImage.filter(ImageFilter.BLUR)
blurImage.show()
```

Boxblurring Image

```python
boxImage = OriImage.filter(ImageFilter.BoxBlur(5))
boxImage.show()
```

Gaussian Blurring image

```python
gaussImage = OriImage.filter(ImageFilter.GaussianBlur(5))
gaussImage.show()
```
